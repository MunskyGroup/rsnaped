# Extracting shape from real images

```python
import os
# rSNAPsim Image Processing
# current working directory
cwd = os.getcwd()
# move one directory up
os.chdir('..')
#importing rSNAPsim.
import rSNAPsim_IP as rss_IP
# return to the working directory
os.chdir(cwd)   
# Plotting
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.path as mpltPath
# To manipulate arrays
import numpy as np 
# skimage library
from skimage import io 
# to save image as tif that can be read by imagej
import tifffile
```

## Extracting shape from CAP-IRES images. Image taken by Amanda.

```python
# %% Section that loads the video and creates the mask that is used a shape for the simulation.
## Loading video
figName = '../Experimental_Data_Videos/CAP_IRES/MAX_Cell01.tif'
## Dimensions for the video
video = io.imread(figName) 
## Using a bandpass filter
low_pass = 1 # low pass filter threshold
high_pass = 251 # high pass filter threshold
temp_obj_BP_filter = rss_IP.Bandpass_filter(video[0:1,:,:,:], low_pass, high_pass)
video_BP_filtered = temp_obj_BP_filter.apply_bp_filter()
## Class: MaskAutomatic
temp_obj = rss_IP.MaskAutomatic(video_BP_filtered,0,0,0,1)
video_removed_mask,mask_array = temp_obj.make_mask()
## Saving shape
tifffile.imwrite('./cell_shape_1.tif', mask_array[0,:,:])
```
