```python
import os
# rSNAPsim Image Processing
# current working directory
cwd = os.getcwd()
# move one directory up
os.chdir('..')
#importing rSNAPsim.
import rSNAPsim_IP as rss_IP
# return to the working directory
os.chdir(cwd)   
# Plotting
import matplotlib as mpl 
import matplotlib.pyplot as plt 
import matplotlib.path as mpltPath
# To manipulate arrays
import numpy as np 
# skimage library
from skimage import io 
# to save image as tif that can be read by imagej
import tifffile

from skimage.filters import sobel
```

```python
# %% Section that loads the video and creates the mask that is used a shape for the simulation.
## Loading video
figName = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Experimental_Data_Videos/CAP_IRES/MAX_Cell01.tif'
## Dimensions for the video
video = io.imread(figName) 
```

```python
video.shape
```

```python
img = video[0,:,:,0].copy()
sobel_img = sobel(img)
plt.imshow(sobel_img,cmap=plt.cm.gray)
plt.rcParams['figure.figsize'] = [12, 12]
```

```python
import cv2
#Autocanny
sigma = 0.3
median = np.median(img)

# apply automatic Canny edge detection using the computed median
lower = int(max(0, (1.0 - sigma) * median))
upper = int(min(255, (1.0 + sigma) * median))

img2 = img.copy()
img2 = np.uint8(img)

auto_canny = cv2.Canny(img2,lower,upper)

cv2.imshow("Auto Canny", auto_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(img,cmap=plt.cm.gray)
plt.rcParams['figure.figsize'] = [15, 15]
```

```python
from skimage.filters import unsharp_mask

#sharpened = unsharp_mask(image0, radius=1.0, amount=1.0)
unsharped_img = unsharp_mask(img, radius=3, amount=1)

plt.imshow(unsharped_img,cmap=plt.cm.gray)
plt.rcParams['figure.figsize'] = [15, 15]

## Class: MaskAutomatic
#temp_obj = rss_IP.MaskAutomatic(unsharped_img,time_point_selected=0, selected_channel=0,mask_all_timePoints=0,use_bandpass_fileter= 0,bandpass_min=0,bandpass_max=0, show_plot=1)
#video_removed_mask,mask_array = temp_obj.make_mask()
```

## Extracting shape from CAP-IRES images. Image taken by Amanda.

```python
# %% Section that loads the video and creates the mask that is used a shape for the simulation.
## Loading video
figName = '../Experimental_Data_Videos/CAP_IRES/MAX_Cell01.tif'
## Dimensions for the video
video = io.imread(figName) 
## Using a bandpass filter
low_pass = 1 # low pass filter threshold
high_pass = 251 # high pass filter threshold
temp_obj_BP_filter = rss_IP.Bandpass_filter(video[0:1,:,:,:], low_pass, high_pass)
video_BP_filtered = temp_obj_BP_filter.apply_bp_filter()
## Class: MaskAutomatic
temp_obj = rss_IP.MaskAutomatic(video_BP_filtered,0,0,0,1)
video_removed_mask,mask_array = temp_obj.make_mask()
## Saving shape
tifffile.imwrite('./cell_shape_1.tif', mask_array[0,:,:])
```
