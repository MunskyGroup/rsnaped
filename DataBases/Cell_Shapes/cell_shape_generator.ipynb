## Extracting shape from CAP-IRES images. Image taken by Amanda.

```python
# %% Section that loads the video and creates the mask that is used a shape for the simulation.
## Loading video
figName = '/Users/luisaguilera/Documents/image_processing_toolbox/DataSets/Cell_Videos/CAP_IRES/MAX_Cell01.tif'
## Dimensions for the video
video = io.imread(figName) 
## Using a bandpass filter
low_pass = 10 # low pass filter threshold
high_pass = 300 # high pass filter threshold
temp_obj_BP_filter = rss_IP.Bandpass_filter(video[1:10,:,:,:], low_pass, high_pass)
video_BP_filtered = temp_obj_BP_filter.apply_bp_filter()
## Class: MaskAutomatic
temp_obj = rss_IP.MaskAutomatic(video_BP_filtered,0,0,0,1)
video_removed_mask,mask_array = temp_obj.make_mask()
## Saving shape
tifffile.imwrite('./cell_shape_0.tif', mask_array[0,:,:])
```
