```python
# Loading modules
import numpy as np
#import time, os, sys
import matplotlib.pyplot as plt; import matplotlib as mpl
from skimage import io ; from skimage.io import imread
%matplotlib inline
mpl.rcParams['figure.dpi'] = 300
```

Single plane images can read into data as nY x nX x channels or channels x nY x nX. 

Multiplane images should read into data as nplanes x channels x nY x nX or as nplanes x nY x nX.

```python
# Loading figures
figName0 = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Project_ONeil/For_Luis/201015_Hela_CTG_E10_CA_#1.tif'
figName1 = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Project_ONeil/For_Luis/201005_Hela_CTG_E10_#2.tif'
figName2 = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Project_ONeil/For_Luis/200926_Hela_CTG_E10_#3.tif'
figName3 = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Project_ONeil/For_Luis/200926_Hela_CTG_E10_#4.tif'
figName4 = '/Users/luisaguilera/Desktop/ImageProcessing_LargeData/Project_ONeil/For_Luis/201015_Hela_CTG_E10_CA_#3.tif'


# Creating a list of figures
files = [figName0]
#files = [figName0, figName1,figName2,figName3,figName4]

imgs = [imread(f) for f in files]
nimg = len(imgs)


imgs[0] = np.transpose(imgs[0], (0, 3, 1, 2))    
#imgs[1] = np.transpose(imgs[1], (0, 3, 1, 2))   
#imgs[2] = np.transpose(imgs[2], (0, 3, 1, 2))   
#imgs[3] = np.transpose(imgs[3], (0, 3, 1, 2))
#imgs[4] = np.transpose(imgs[4], (0, 3, 1, 2))   

imgs[0] = imgs[0][:,0:3,:,:]
#imgs[1] = imgs[1][:,0:3,:,:]
#imgs[2] = imgs[2][:,0:3,:,:]
#imgs[3] = imgs[3][:,0:3,:,:]
#imgs[4] = imgs[4][:,0:3,:,:]
print(imgs[0].shape)

#plt.figure(figsize=(12,4))
#for k,img in enumerate(imgs):
#    plt.subplot(1,nimg,k+1)
#    plt.imshow(img)
```

```python
# RUN CELLPOSE

from cellpose import models, io

# DEFINE CELLPOSE MODEL
# model_type='cyto' or model_type='nuclei'
model = models.Cellpose(gpu=False, model_type='cyto')

# define CHANNELS to run segementation on
# grayscale=0, R=1, G=2, B=3
# channels = [cytoplasm, nucleus]
# if NUCLEUS channel does not exist, set the second channel to 0
# channels = [0,0]
# IF ALL YOUR IMAGES ARE THE SAME TYPE, you can give a list with 2 elements
# channels = [0,0] # IF YOU HAVE GRAYSCALE
# channels = [2,3] # IF YOU HAVE G=cytoplasm and B=nucleus
# channels = [2,1] # IF YOU HAVE G=cytoplasm and R=nucleus

# or if you have different types of channels in each image
#channels = [[2,1],[2,1],[2,1],[2,1]]
channels = [2,1]


# if diameter is set to None, the size of the cells is estimated on a per image basis
# you can set the average cell `diameter` in pixels yourself (recommended) 
# diameter can be a list or a single number for all images
masks, flows, styles, diams = model.eval(imgs, diameter=100, do_3D=True)
```

```python
# DISPLAY RESULTS
from cellpose import plot

nimg = len(imgs)
for idx in range(nimg):
    maski = masks[idx]
    flowi = flows[idx][0]
    
    fig = plt.figure(figsize=(12,5))
    plot.show_segmentation(fig, imgs[idx], maski, flowi, channels=channels[idx])
    plt.tight_layout()
    plt.show()
```
