## Stochastic simulations - rSNAPsim

```python
##################################
##################################
## Stochastic Simulations
number_spots = 1
n_frames = 100 # in seconds
step_size = 1 # in seconds
##################################
##################################
```

```python
# %% Section that runs the stochastic simulations
# Loading modules
rSNAPsim_directory = '/Users/luisaguilera/Documents/rSNAPsim-dev/'
sequences_directory = '/Users/luisaguilera/Documents/rSNAPsim-dev/gene_files/Bactin_withTags.txt'
cwd = os.getcwd(); os.chdir(rSNAPsim_directory); import rSNAPsim; rss = rSNAPsim.rSNAPsim(); os.chdir(cwd); rss = rSNAPsim.rSNAPsim()
# RUNNING THE SSA
rss.open_seq_file(sequences_directory)
rss.get_orfs(rss.sequence_str) # now to look for open reading frames
rss.orfs # reading the 3 ORF.
protein1 = rss.sequence_str # Deffining a Protein passing the DNA sequence  
rss.nt2aa(protein1) # Converting nucleotides to Protein
rss.get_temporal_proteins()  #proteins sorted by
rss.analyze_poi(rss.pois[0],rss.pois_seq[0]) # Analyze the first protein, analyze function takes the aa_seq then nt_seq
rss.POI 
# Running the SSA solver 
starting_time = 1000
ssa1 = rss.ssa_solver(n_traj = number_spots, start_time=starting_time,tf=starting_time+n_frames, tstep=starting_time+n_frames,k_elong_mean=3, k_initiation=.05)  # tstep = total number of steps including the burnin time 
simulated_trajectories = ssa1.intensity_vec # Vector that will be passed to the simulated cell
ssa2 = rss.ssa_solver(n_traj = number_spots, start_time=starting_time,tf=starting_time+n_frames, tstep=starting_time+n_frames,k_elong_mean=3, k_initiation=.05)  # tstep = total number of steps including the burnin time 
simulated_trajectories_blue = ssa2.intensity_vec # Vector that will be passed to the simulated cell
```

## <span style= "color:red"> Simulated Cell </span>:  Shape+Noise+Cytoskeleton+Photobleaching

```python
# Parameter definifition for the simulation.

# directly load a cell shape
shape_directory = '../cell_shapes/cell_shape_2.tif'
cell_shape = io.imread(shape_directory) 

# Parameter for the cell
contours = np.array(find_contours(cell_shape, 0.8),dtype = int)
polygon_array = contours[0]  # defauls is a circle
width = cell_shape.shape[0]
height = cell_shape.shape[1]
image_size = [width,height,1]
diffusion_coeffient = 1 # Diffusion coeffient for Brownian Motion

# Paramerters for channel 1 -> RED
minimal_background_outside_cell_ch1 =10
minimal_background_inside_cell_ch1 =40
size_spot_ch1 = 5
spot_sigma_ch1 = 1.5

add_background_noise_ch1 = 1
amount_background_noise_ch1 = 0.005

add_photobleaching_ch1 = 0
photobleaching_exp_constant_ch1 = 0.0005
simulated_trajectories_ch1 =0
ignore_ch1 = 0
use_triangulation_ch1 =0

max_phoptobleaching = np.log(5)/ n_frames # this line defines the maximum value for the exponential constant that doesnt excedes 255(max uint8 value). 
# Paramerters for channel 2 -> GREEN
minimal_background_outside_cell_ch2 =10
minimal_background_inside_cell_ch2 =40
size_spot_ch2 = 5
spot_sigma_ch2 = 1.5
add_background_noise_ch2 = 1
amount_background_noise_ch2 = 0.001
add_photobleaching_ch2 = 0
photobleaching_exp_constant_ch2 = max_phoptobleaching*0.5
simulated_trajectories_ch2 = simulated_trajectories
ignore_ch2 = 0
use_triangulation_ch2=0

# Paramerters for channel 3 -> BLUE
minimal_background_outside_cell_ch3 =10
minimal_background_inside_cell_ch3 =40
size_spot_ch3 = 5
spot_sigma_ch3 = 1.5
add_background_noise_ch3 = 1
amount_background_noise_ch3 = 0.001
add_photobleaching_ch3 = 1
photobleaching_exp_constant_ch3 = max_phoptobleaching*0.9
simulated_trajectories_ch3 = simulated_trajectories_blue
ignore_ch3 = 0
use_triangulation_ch3 = 0
```

```python
# Running the simulation
tensor_video, tensor_for_image_j, spot_positions_movement, tensor_mean_intensity_in_figure, tensor_std_intensity_in_figure = rss_IP.Simulated_cell(number_spots, n_frames, step_size, diffusion_coeffient, polygon_array, image_size, simulated_trajectories_ch1, minimal_background_outside_cell_ch1, minimal_background_inside_cell_ch1, size_spot_ch1, spot_sigma_ch1, add_background_noise_ch1, amount_background_noise_ch1, add_photobleaching_ch1, photobleaching_exp_constant_ch1, simulated_trajectories_ch2, minimal_background_outside_cell_ch2, minimal_background_inside_cell_ch2, size_spot_ch2, spot_sigma_ch2, add_background_noise_ch2, amount_background_noise_ch2, add_photobleaching_ch2, photobleaching_exp_constant_ch2, simulated_trajectories_ch3, minimal_background_outside_cell_ch3, minimal_background_inside_cell_ch3, size_spot_ch3, spot_sigma_ch3, add_background_noise_ch3, amount_background_noise_ch3, add_photobleaching_ch3, photobleaching_exp_constant_ch3, ignore_ch1,ignore_ch2, ignore_ch3,use_triangulation_ch1,use_triangulation_ch2,use_triangulation_ch3).make_simulated_cell()
```

### Comparing positions - 2D-Brownian trajectories

```python
fig, ax = plt.subplots(1,3, figsize=(15, 3))
if number_detected_particles == 1:
    #Initial points
    ax[0].plot(spot_positions_movement[0,0,0],spot_positions_movement[0,0,1],'o',linewidth = 3, markersize=20,color='b')
    ax[1].plot(DataFrame_intensities[DataFrame_intensities['Spot_No'] ==1].y_position.values[0],DataFrame_intensities[DataFrame_intensities['Spot_No'] ==1].x_position.values[0],'o',linewidth = 3, markersize=20,color='b')
    #End points
    ax[0].plot(spot_positions_movement[-1,0,0],spot_positions_movement[-1,0,1],'>',linewidth = 3, markersize=20,color='b')
    ax[1].plot(DataFrame_intensities[DataFrame_intensities['Spot_No'] ==1].y_position.values[-1],DataFrame_intensities[DataFrame_intensities['Spot_No'] ==1].x_position.values[-1],'>',linewidth = 3, markersize=20,color='b')
# Real positions in simulated cell - from tensor: spot_positions_movement
for tr  in range(1,number_detected_particles+1):
    ax[0].plot(spot_positions_movement[:,tr-1,0],spot_positions_movement[:,tr-1,1],'o',linewidth = 3, markersize=10,color='k')
    ax[0].set(xlabel='x-position')
    ax[0].set(ylabel='y-position')
    ax[0].set(title='Real (tensor: spot_positions_movement)');
# Spots detection using trackpy    
for tr in range(1,number_detected_particles+1):
    ax[1].plot(DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr].y_position.values,DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr].x_position.values,'-',linewidth = 3, markersize=10,color='m')
ax[1].set(xlabel='x-position')
ax[1].set(ylabel='y-position')
ax[1].set(title='Localization using trackpy')

# Comparing Spots in simulated cell vs detection with trackpy
for tr in range(1,number_detected_particles+1):    
    ax[2].plot(spot_positions_movement[:,tr-1,0],spot_positions_movement[:,tr-1,1],'o',linewidth = 3, markersize=10,color='k')
for tr in range(1,number_detected_particles+1):
    ax[2].plot(DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr].y_position.values,DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr].x_position.values,'-',linewidth = 3, markersize=10,color='m')
ax[2].set(xlabel='x-position')
ax[2].set(ylabel='y-position')
ax[2].set(title='Overlapped positions')
plt.show()
```

### Intensity in simulated cell vs SSA

```python
plt.figure(figsize=(15,2))
for tr in range(0,ssa1.n_traj ):
    plt.plot(ssa1.time_rec - starting_time , ssa1.intensity_vec[tr,:] , '-',linewidth = 3,color='green');
    tmp_int = DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr+1].Green_Int.values
    tmp_time = DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr+1].Frame.values
    plt.plot(tmp_time,tmp_int , 'o',linewidth = 3, markersize=7,color='m')
plt.legend(["SSA", "Calculated"], loc ="top right") 
plt.title('Intensities (a.u.)')
plt.show()
```

### Normalized intensity in simulated cell vs SSA

```python
def normalization(vector):
    # removing min and normalizing by maximum
    norm_vector = vector - np.amin(vector)
    norm_vector = norm_vector/(np.amax(norm_vector)*0.99)
    return norm_vector

plt.figure(figsize=(20,3))
for tr in range(0,ssa1.n_traj):
    #plt.plot(ssa1.time_rec - starting_time , ssa1.intensity_vec[tr,:]/np.amax(ssa1.intensity_vec[tr,:]) , '-',linewidth = 3,color='green');
    plt.plot(ssa1.time_rec - starting_time ,  normalization(ssa1.intensity_vec[tr,:]), '-',linewidth = 3,color='green');

    try:
        tmp_int = DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr+1].Green_Int.values
        tmp_time = DataFrame_intensities[DataFrame_intensities['Spot_No'] ==tr+1].Frame.values
        #plt.plot(tmp_time,tmp_int/np.amax(tmp_int)*1.05, 'o',linewidth = 3, markersize=7,color='m')
        plt.plot(tmp_time, normalization(tmp_int), '-o',linewidth = 3, markersize=7,color='m')
    except:
        pass
plt.title('Normalized intensities (a.u.)')
plt.legend(["SSA", "Calculated"], loc ="top right") 
plt.show()
```
